```r
# Dropping unnecessary columns
mod_data <- data %>%
  select(-url, -address, -phone)
```

```r
# Removing duplicates
mod_data <- mod_data %>%
  distinct()
```

```r
# Dropping rows with missing values
mod_data <- mod_data %>%
  drop_na()

# Renaming columns to remove spaces and periods
names(mod_data) <- make.names(names(mod_data))
```

```r
library(dplyr)
mod_data <- rename(mod_data, cost = approx_cost.for.two.people.)
mod_data <- rename(mod_data, type = listed_in.type.)
mod_data <- rename(mod_data, city = listed_in.city.)
```

```r
print(colnames(mod_data))
```

```r
mod_data$cost <- as.character(mod_data$cost)
mod_data$cost <- gsub(",", ".", mod_data$cost)
mod_data$cost <- as.numeric(mod_data$cost)
print(str(mod_data))
```

```r
mod_data$name <- sapply(mod_data$name, function(x) {
  paste0(tolower(substring(x, 1, 1)), substring(x, 2))
})
mod_data$online_order <- ifelse(mod_data$online_order == 'Yes', TRUE, FALSE)
mod_data$book_table <- ifelse(mod_data$book_table == 'Yes', TRUE, FALSE)
```

```r
# Converting 'rate' column to numeric by extracting numerical values
mod_data$rate <- as.numeric(gsub(".*?([0-9.]+).*", "\\1", mod_data$rate))

# Removing '/5' from votes column and convert to numeric
mod_data$votes <- as.numeric(gsub("/5", "", mod_data$votes))
```

```r
# Get unique restaurant names
restaurants <- unique(mod_data$name)
mod_data$Mean.Rating <- 0
for (i in 1:length(restaurants)) {
  index <- mod_data$name == restaurants[i]
  mod_data$Mean.Rating[index] <- mean(mod_data$rate[index], na.rm = TRUE)
}
```

```r
# Load required library
library(scales)
mod_data$Mean.Rating <- rescale(mod_data$Mean.Rating, to = c(1, 5))
mod_data$Mean.Rating <- round(mod_data$Mean.Rating, 2)
sample_n(mod_data, 3)
```

```r
stopwords <- tm::stopwords("en")
remove_stopwords <- function(text) {
  words <- strsplit(text, "\\s")[[1]]
  words <- words[!words %in% stopwords]
  return(paste(words, collapse = " "))
}
mod_data$reviews_list <- sapply(mod_data$reviews_list, remove_stopwords)
```

```r
library(stringr)
remove_urls <- function(text) {
  url_pattern <- "https?://[^\\s]+|www\\.[^\\s]+"
  return(str_remove_all(text, url_pattern))
}
mod_data$reviews_list <- sapply(mod_data$reviews_list, remove_urls)
```

```r
get_top_words <- function(column, top_nu_of_words, nu_of_word) {
  corpus <- Corpus(VectorSource(column))
  corpus <- tm_map(corpus, content_transformer(tolower))
  corpus <- tm_map(corpus, removePunctuation)
  corpus <- tm_map(corpus, removeWords, stopwords("en"))
  dtm <- DocumentTermMatrix(corpus)
  freq <- colSums(as.matrix(dtm))
  words_freq <- sort(freq, decreasing = TRUE)
  return(words_freq[1:top_nu_of_words])
}
```

```r
library(tm)
library(Matrix)
corpus <- Corpus(VectorSource(df_percent$reviews_list))
corpus <- tm_map(corpus, content_transformer(tolower))
corpus <- tm_map(corpus, removeNumbers)
corpus <- tm_map(corpus, removePunctuation)
corpus <- tm_map(corpus, removeWords, stopwords("en"))
corpus <- tm_map(corpus, stripWhitespace)
dtm <- DocumentTermMatrix(corpus)
tfidf <- weightTfIdf(dtm)
```

```r
library(Matrix)
tfidf_sparse <- as(Matrix::Matrix(tfidf_matrix, sparse = TRUE), "Matrix")
cosine_similarities <- Matrix::crossprod(tfidf_sparse) / sqrt(Matrix::crossprod(tfidf_sparse) * Matrix::crossprod(tfidf_sparse))
```

```r
recommend <- function(name, cosine_similarities, indices, df_percent) {
  recommend_restaurant <- vector("list", length = 10)
  idx <- which(indices == name)
  if (length(idx) > 0) {
    idx <- idx[1]
    score_series <- sort(cosine_similarities[idx, ], decreasing = TRUE)
    top30_indexes <- head(names(score_series), 31)
    for (each in top30_indexes) {
      recommend_restaurant <- c(recommend_restaurant, list(df_percent[each, c("name", "Mean.Rating"), drop = FALSE]))
    }
    df_new <- do.call(rbind, recommend_restaurant)
    df_new <- df_new[!duplicated(df_new), ]
    df_new <- df_new[order(-df_new$Mean.Rating), ][1:10, ]
    cat(paste("TOP", nrow(df_new), "RESTAURANTS LIKE", name, "WITH SIMILAR REVIEWS:\n"))
    print(df_new[, c("name", "Mean.Rating")])
    return(df_new[, c("name", "Mean.Rating")])

  } else {
    cat("Recommending top 10 restaurants with similar reviews:\n")
    top10_indexes <- head(order(-rowSums(cosine_similarities), decreasing = TRUE), 10)
    for (each in top10_indexes) {
      recommend_restaurant <- c(recommend_restaurant, list(df_percent[each, c("name", "Mean.Rating"), drop = FALSE]))
    }
    df_new <- do.call(rbind, recommend_restaurant)
    df_new <- df_new[!duplicated(df_new), ]
    print(df_new[, c("name", "Mean.Rating")])

    return(df_new[, c("name", "Mean.Rating")])
  }
}

recommend_restaurants <- recommend("Pai Vihar", cosine_similarities, indices, df_percent)
print(recommend_restaurants)
```
